```python
# importing 
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store CSV into DataFrame
csv_file = "data/netflix_titles.csv"
netflix_data_df = pd.read_csv(csv_file)
netflix_data_df.head()
```

```python
netflix_data_df.count()
```

```python
# Dropping every record which has NaN value in it. 
nf_na_df = netflix_data_df.dropna()
```

```python
nf_na_df.count()
```

```python
# Filtering only 'United State' 
nf_us_df = nf_na_df.loc[nf_na_df["country"] == "United States", :]
```

```python
only_movies_df = nf_us_df.loc[nf_us_df["type"] == "Movie", :]
```

```python
only_movies_df.head()
```

```python
only_movies_df.count()
```

```python
only_movies_df.dtypes
```

```python
# Dropping column "country" due to repeated value and "date_added"
only_col_df = only_movies_df.drop(columns=['type','country','date_added'])
```

```python
only_col_df.head()
```

```python
# Dropping duplicate titles 
dp_df = only_col_df.drop_duplicates(subset=['title'], keep='last')
```

```python
# Removing "min" from "duration" column value and change data type to as float
dp_df[dp_df.columns[6]] = dp_df[dp_df.columns[6]].replace('[min]', '', regex=True).astype(float)

```

```python
# CREATING COPY OF EXISTING DF AS SAFE DATA POINT FOR ANY FUTURE CHANGES.
clean_netflix_df = dp_df.copy()
```
