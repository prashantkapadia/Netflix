```python
# importing 
import pandas as pd
from sqlalchemy import create_engine
```

```python
streaming_data = "final_stream.csv"
streaming_data_df = pd.read_csv(streaming_data)
```

```python
connection_string = "admin1:12345@localhost:5432/Project#2"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Below code worked and loaded successfully into pgAdmin Database table. 
streaming_data_df.to_sql(name='netflix_stream', con=engine, if_exists='append', index=False)
```
